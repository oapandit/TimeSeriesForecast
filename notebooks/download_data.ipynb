```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the ticker symbols for each asset
tickers = ["^BSESN", "^NSEI", "GOLD", "^GSPC"]

# Define the date range for historical data (from 2005 to 2023)
start_date = "2005-01-01"
end_date = "2023-01-01"

# Function to fetch historical data for a specific year
def fetch_data_for_year(tickers, year):
    start_date = f"{year}-01-01"
    end_date = f"{year}-12-31"
    data = yf.download(tickers, start=start_date, end=end_date)["Adj Close"]
    return data

# Function to calculate correlation matrix for a specific year
def calculate_correlation_matrix(data):
    return data.corr().values.flatten()

# Initialize lists to store correlation values and years
years = range(2005, 2024)
correlation_values = []

# Loop through each year from 2005 to 2023, calculate correlation matrix
for year in years:
    # Fetch historical data for the year
    historical_data = fetch_data_for_year(tickers, year)
    
    # Calculate correlation matrix for the year and flatten to 1D array
    corr_values = calculate_correlation_matrix(historical_data)
    
    # Append to correlation_values list
    correlation_values.append(corr_values)

# Convert correlation_values to DataFrame for easy plotting
correlation_df = pd.DataFrame(correlation_values, columns=["BSESN_NSEI", "BSESN_GOLD", "BSESN_GSPC", "NSEI_GOLD", "NSEI_GSPC", "GOLD_GSPC"])

# Plotting line plot of correlation changes over time
plt.figure(figsize=(12, 8))
for column in correlation_df.columns:
    plt.plot(years, correlation_df[column], marker='o', label=column)

plt.title('Yearly Correlation Changes')
plt.xlabel('Year')
plt.ylabel('Correlation')
plt.xticks(years, rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the ticker symbols for each asset
tickers = ["^BSESN", "^NSEI", "GOLD", "^GSPC"]

# Define the date range for historical data (from 2005 to 2023)
start_date = "2005-01-01"
end_date = "2023-01-01"

# Function to fetch historical data for a specific year
def fetch_data_for_year(tickers, year):
    start_date = f"{year}-01-01"
    end_date = f"{year}-12-31"
    data = yf.download(tickers, start=start_date, end=end_date)["Adj Close"]
    return data

# Function to calculate correlation matrix for a specific year
def calculate_correlation_matrix(data):
    return data.corr()

# Initialize an empty DataFrame to store yearly correlation matrices
correlation_matrices = {}

# Loop through each year from 2005 to 2023, calculate correlation matrix
for year in range(2005, 2024):
    # Fetch historical data for the year
    historical_data = fetch_data_for_year(tickers, year)
    
    # Calculate correlation matrix for the year
    correlation_matrix = calculate_correlation_matrix(historical_data)
    
    # Store correlation matrix in dictionary
    correlation_matrices[year] = correlation_matrix

# Plotting heatmap of yearly correlation matrices
plt.figure(figsize=(12, 8))
for year in range(2005, 2024):
    plt.subplot(4, 5, year - 2004)  # Adjust subplot grid for the years
    sns.heatmap(correlation_matrices[year], annot=True, cmap="coolwarm", vmin=-1, vmax=1)
    plt.title(f'{year}')
    plt.tight_layout()

plt.suptitle('Yearly Correlation Matrix Heatmaps', y=1.02)
plt.show()
```
