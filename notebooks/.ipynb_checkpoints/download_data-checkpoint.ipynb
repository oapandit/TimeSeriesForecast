```python
import yfinance as yf
import pandas as pd

# Define the ticker symbol for the stock
ticker_symbol = "RELIANCE.NS"  # Example: Reliance Industries on NSE

# Set the start date and end date for the historical data
start_date = "1995-01-01"
end_date = "2024-01-01"

# Fetch historical data from Yahoo Finance
data = yf.download(ticker_symbol, start=start_date, end=end_date)

# Print the first few rows of the dataframe
print(data.head())

# Save the data to a CSV file
data.to_csv(f"{ticker_symbol}_historical_data.csv")
```

```python
import yfinance as yf
import pandas as pd

# Define the ticker symbols for each asset
tickers = ["^BSESN", "^NSEI", "GOLD", "^GSPC"]

# Define the date range for historical data
start_date = "2019-01-01"
end_date = "2023-01-01"

# Function to fetch historical data
def fetch_data(tickers, start_date, end_date):
    data = yf.download(tickers, start=start_date, end=end_date)["Adj Close"]
    return data

# Fetch historical data
historical_data = fetch_data(tickers, start_date, end_date)

# Display the first few rows of the data
print(historical_data.head())

# Calculate correlation matrix
correlation_matrix = historical_data.corr()

# Display correlation matrix
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title('Correlation Matrix of Sensex, Nifty, Gold Price, S&P 500')
plt.show()
```
